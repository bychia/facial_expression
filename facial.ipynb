```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def image_converted_to_gray(image):
    # Converting to grayscale
    image_converted_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Displaying the grayscale image
    plt.imshow(image_converted_gray, cmap='gray')
    return image_converted_gray
    
```

```python
image = cv2.imread("images/20191022_181144_994.jpg")
image_gray = image_converted_to_gray(image)
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(image_gray, scaleFactor = 1.2, minNeighbors = 5);
```

```python
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 5)
```

```python
plt.imshow(convertToRGB(image))
```
